```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import gc
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from xgboost import XGBClassifier
filepath = r'https://media.githubusercontent.com/media/mohit12kumar/dataset4/refs/heads/main/project14.csv'
data = pd.read_csv(filepath)
data.head()
```

```python
x= data[[
    'AIRLINE',
    'ORIGIN',
    'DEST',
    'DEP_DELAY',
    'DISTANCE',
    'CRS_DEP_TIME',
    'month',
    'day_of_week',
    'is_weekend'
]]
y = (data["ARR_DELAY"] > 15).astype(int)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42 )
# sm = SMOTE(random_state=42)
# x_train, y_train = sm.fit_resample(x_train, y_train)
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=15,
    min_samples_split=10,
    min_samples_leaf=5,
    class_weight='balanced',
    n_jobs=2,
    random_state=42
)

rf.fit(x_train, y_train)
rf_prob = rf.predict_proba(x_test)[:,1]

threshold = 0.30
rf_pred = (rf_prob > threshold).astype(int)
print(rf_pred, y_test.values)

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

print("\n test accuracy = ", accuracy_score(y_test, rf_pred))
print("Report:\n", classification_report(y_test, rf_pred))

c1 = confusion_matrix(y_test, rf_pred)
print("confusion matrix:\n", c1)

unique_labels = sorted(list(set(y_test).union(set(rf_pred))))

cm_display = ConfusionMatrixDisplay(confusion_matrix=c1, display_labels=unique_labels)
cm_display.plot(cmap="Blues", xticks_rotation=45)
plt.show()

import gc
gc.collect()
```

```python
from sklearn.svm import LinearSVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import gc

scaler = StandardScaler()

x_trains = scaler.fit_transform(x_train)
x_tests = scaler.transform(x_test)

svm = LinearSVC(
    class_weight='balanced',
    max_iter=3000,
    dual=False,
    random_state=42
)

svm.fit(x_trains, y_train)

svm_scores = svm.decision_function(x_tests)

threshold = -0.2

svm_pred = (svm_scores > threshold).astype(int)

print(svm_pred, y_test.values)

print("\n test accuracy = ", accuracy_score(y_test, svm_pred))
print("Report:\n", classification_report(y_test, svm_pred))

c5 = confusion_matrix(y_test, svm_pred)
print("confusion matrix:\n", c5)

unique_labels = sorted(list(set(y_test).union(set(svm_pred))))

cm_display = ConfusionMatrixDisplay(
    confusion_matrix=c5,
    display_labels=unique_labels
)

cm_display.plot(cmap="Blues", xticks_rotation=45)
plt.show()

gc.collect()
```

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from sklearn.utils.class_weight import compute_class_weight

scaler = StandardScaler()

x_trains = scaler.fit_transform(x_train)
x_tests = scaler.transform(x_test)


classes = np.unique(y_train)

weights = compute_class_weight(
    class_weight='balanced',
    classes=classes,
    y=y_train
)

class_weights = dict(zip(classes, weights))
print("Class Weights:", class_weights)


model = Sequential([
    Dense(128, activation='relu', input_dim=x_trains.shape[1]),
    BatchNormalization(),
    Dropout(0.3),

    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    Dense(32, activation='relu'),
    BatchNormalization(),

    Dense(1, activation='sigmoid')
])

optimizer = Adam(learning_rate=0.001)

model.compile(
    optimizer=optimizer,
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

history = model.fit(
    x_trains, y_train,
    epochs=50,
    batch_size=64,
    validation_data=(x_tests, y_test),
    callbacks=[early_stop],
    class_weight=class_weights
)


dL_prob = model.predict(x_tests)

threshold = 0.30
dL_pred = (dL_prob > threshold).astype(int)

print(dL_pred, y_test.values)
```

```python
results = []

def evaluate_model(name, y_true, y_pred):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred),
        "Recall": recall_score(y_true, y_pred),
        "F1": f1_score(y_true, y_pred)
    })
evaluate_model("Random Forest", y_test, rf_pred)
evaluate_model("XGBoost", y_test, xg_pred)
evaluate_model("Deep Learning", y_test, dL_pred)
evaluate_model("KNN", y_test, knn_pred)
evaluate_model("Logistic Regression", y_test, lr_pred)
evaluate_model("decision tree", y_test, dl_pred)
evaluate_model("svm", y_test, svm_pred)
```

```python
import pandas as pd
import numpy as np
import joblib
import warnings

warnings.filterwarnings("ignore")

# ==============================
# 🔥 DL CHECK
# ==============================
try:
    import tensorflow as tf
    from tensorflow.keras.models import load_model
    DL_AVAILABLE = True
except:
    DL_AVAILABLE = False

# ==============================
# 🔥 LOAD FILES
# ==============================
encoders = joblib.load("encoders.pkl")

try:
    scaler = joblib.load("scaler_delay.pkl")
except:
    scaler = None

# ==============================
# 🔥 LOAD MODELS
# ==============================
models = {
    "Random Forest": joblib.load("random_forest_model_DELAY.pkl"),
    "Decision Tree": joblib.load("decision_tree_model_DELAY.pkl"),
    "Logistic Regression": joblib.load("LogisticRegression_DELAY.pkl"),
    "SVM": joblib.load("LinearSVC_DELAY.pkl"),
    "KNN": joblib.load("KNeighborsClassifier_DELAY.pkl"),
    "XGBoost": joblib.load("XGBClassifier
    .pkl")
}

if DL_AVAILABLE:
    models["Deep Learning"] = load_model("deep_learning_DELAY.keras", compile=False)

# ==============================
# 🔥 GLOBAL THRESHOLD
# ==============================
THRESHOLD = 0.30

# ==============================
# 🔥 DL FIX (NO RETRACING)
# ==============================
if DL_AVAILABLE:
    dl_model = models["Deep Learning"]

    @tf.function(reduce_retracing=True)
    def dl_predict(x):
        return dl_model(x, training=False)

# ==============================
# 🔥 SAFE ENCODER
# ==============================
def safe_encode(le, value):
    if value in le.classes_:
        return le.transform([value])[0]
    return le.transform([le.classes_[0]])[0]

# ==============================
# 🔥 INPUT
# ==============================
airline = input("Airline: ").strip().upper()
origin = input("Origin: ").strip().upper()
dest = input("Destination: ").strip().upper()

dep_delay = max(float(input("Departure Delay: ")), 0)
distance = float(input("Distance: "))
time = float(input("CRS Time: "))

month = int(input("Month (1-12): "))
day = int(input("Day (1-7): "))
is_weekend = 1 if day in [6,7] else 0

# ==============================
# 🔥 ENCODING
# ==============================
airline_enc = safe_encode(encoders["AIRLINE"], airline)
origin_enc = safe_encode(encoders["ORIGIN"], origin)
dest_enc = safe_encode(encoders["DEST"], dest)

# ==============================
# 🔥 DATAFRAME
# ==============================
feature_order = [
    'AIRLINE','ORIGIN','DEST',
    'DEP_DELAY','DISTANCE','CRS_DEP_TIME',
    'month','day_of_week','is_weekend'
]

df = pd.DataFrame([[
    airline_enc, origin_enc, dest_enc,
    dep_delay, distance, time,
    month, day, is_weekend
]], columns=feature_order)

# ==============================
# 🔥 PREDICTION
# ==============================
results = {}

for name, model in models.items():
    temp_df = df.copy()

    try:
        # 🔥 SCALING
        if scaler and name in ["Logistic Regression","SVM","KNN","Deep Learning"]:
            temp_df = pd.DataFrame(
                scaler.transform(temp_df),
                columns=feature_order
            )

        # ======================
        # 🔥 DEEP LEARNING FIX
        # ======================
        if name == "Deep Learning":
            prob = float(dl_predict(temp_df).numpy()[0][0])
            pred = int(prob > THRESHOLD)

        # ======================
        # 🔥 SVM FIX
        # ======================
        elif name == "SVM":
            score = model.decision_function(temp_df)[0]
            pred = int(score > -0.2)

        # ======================
        # 🔥 XGBOOST FIX
        # ======================
        elif name == "XGBoost":
            prob = float(model.predict_proba(temp_df)[0][1])
            pred = int(prob > THRESHOLD)

        # ======================
        # 🔥 OTHER MODELS
        # ======================
        elif hasattr(model, "predict_proba"):
            prob = float(model.predict_proba(temp_df)[0][1])
            pred = int(prob > THRESHOLD)

        # ======================
        # 🔥 FALLBACK
        # ======================
        else:
            pred = int(model.predict(temp_df)[0])

        results[name] = pred

    except Exception as e:
        results[name] = f"Error: {str(e)}"

# ==============================
# 🔥 OUTPUT
# ==============================
print("\n✈️ MODEL RESULTS:\n")

for k, v in results.items():
    if isinstance(v, int):
        print(f"{k}: {'🔴 Delay' if v==1 else '🟢 On Time'}")
    else:
        print(f"{k}: ⚠️ {v}")

# ==============================
# 🔥 FINAL VOTING
# ==============================
votes = [v for v in results.values() if isinstance(v, int)]

if votes:
    final = round(sum(votes)/len(votes))
    print("\n🎯 FINAL RESULT:",
          "🔴 Delay" if final==1 else "🟢 On Time")
```
